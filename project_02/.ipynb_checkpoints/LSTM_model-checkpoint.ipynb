```python
data = pd.read_csv("data/competitive-data-science-predict-future-sales/sales_train_v2.csv")
ts = pd.Series(data['item_price'].values, index=data['date']) 
data = data.iloc[0:20000,4:5]
data.head()
```

```python
rcParams['figure.figsize'] = 18, 6
smt.graphics.plot_acf(data[0:500], lags=20, alpha=0.5)
```

```python
rcParams['figure.figsize'] = 18, 6
plt.plot(data[0:500], color='blue', label='Item price')
plt.legend()
```

```python
data = np.array(data, dtype=np.float32)
data_raw = np.reshape(data, -1)
```

```python
# normaliza
temp = np.array(data_raw).reshape((len(data_raw), 1))
scaler = MinMaxScaler(feature_range=(0, 1))
dataset = scaler.fit_transform(temp)
dataset = np.reshape(dataset, -1)
```

```python
# divide a base de acordo com o tamanho da janela.
def create_dataset(dataset, window_size=1):
    data_X, data_Y = [], []
    for i in range(len(dataset) - window_size - 1):
        a = dataset[i:(i + window_size)]
        data_X.append(a)
        data_Y.append(dataset[i + window_size])
    return(np.array(data_X), np.array(data_Y))
```

```python
print(dataset.shape)
data_size=dataset.shape[0]

# retira parte dos dados para teste
train_size = 2*(data_size//3)
test_size = data_size

print(train_size)
print(test_size)
```

```python
train, test = dataset[0:train_size], dataset[train_size:test_size]
```

```python
# aplica o metodo para criar as janelas para o treino com um tamanho de 20 pontos
window_size = 20
train_X, train_Y = create_dataset(train, window_size)
test_X, test_Y = create_dataset(test, window_size)
```

```python
# redimensiona entrada
train_X = np.reshape(train_X, (train_X.shape[0], 1, train_X.shape[1]))
test_X = np.reshape(test_X, (test_X.shape[0], 1, test_X.shape[1]))
```

```python
# mtodo que cria o modelo para lstm
def lstm_model():
    
    model = Sequential()
    model.add(LSTM(128, return_sequences=True, input_shape=(1, window_size)))
    model.add(Dropout(0.2))
    
    model.add(LSTM(256, return_sequences=False))
    model.add(Dropout(0.2))
    
    model.add(Dense(1))
    model.add(Activation("linear"))
    
    model.compile(loss="mean_squared_error", optimizer="adam", metrics=["mape", "mae"])
    
    model.fit(train_X, train_Y, epochs=80, batch_size=10, verbose=2)
    
    
    # ver se e no de teste ou de treino
    pred_Y = model.predict(test_X)
    
    mse = mean_squared_error(test_Y, pred_Y)
    rmse = math.sqrt(mse)
    mae = mean_absolute_error(test_Y, pred_Y)
    evs = explained_variance_score(test_Y, pred_Y)
    r2s = r2_score(test_Y, pred_Y)
    
    print('MSE: {:.8f} RMSE: {:.8f} MAE: {:.8f} EVS: {:.8f} R2S: {:.8f}'.format(mse, rmse, mae, evs, r2s))

    return pred_Y
```

```python
pred_Y = lstm_model()
```

```python
print('Plotting Results')
rcParams['figure.figsize'] = 18, 6
plt.title('Expected')
plt.plot(test_Y, color='blue', label='Train')
plt.title('Predicted')
plt.plot(pred_Y, color='red', label='Test')
plt.legend()
```

```python
mse = mean_squared_error(test_Y, pred_Y)
rmse = math.sqrt(mse)
mae = mean_absolute_error(test_Y, pred_Y)
evs = explained_variance_score(test_Y, pred_Y)
r2s = r2_score(test_Y, pred_Y)
    
print('MSE: {:.8f} RMSE: {:.8f} MAE: {:.8f} EVS: {:.8f} R2S: {:.8f}'.format(mse, rmse, mae, evs, r2s))
```
